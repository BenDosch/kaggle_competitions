# Preprocess Training Data

```python

# Load training data
train = pd.read_csv("/kaggle/input/titanic/train.csv")


# Check column names and datatypes
print(train.dtypes)

# Drop unwanted columns
train = train.drop(labels=["PassengerId", "Name", "Ticket", "Fare", "Cabin"], axis=1)

# Find columns with Null values.
if train.isnull().values.any():
    for col in train.columns:
        if train[col].isnull().values.any():
            print(col, " has ", train[col].isnull().sum(), " NaN.")
else:
    print("No NaN detected.")

    
# Fill in missing values
train["Age"] = train["Age"].fillna(train["Age"].median())
train["Embarked"] = train["Embarked"].fillna(train["Embarked"].mode())

# Factorize non-numeric
train["Sex"], _ = pd.factorize(train["Sex"])
train["Embarked"], _ = pd.factorize(train["Embarked"])

# Visual check
train.head()
```

Note about data:
SibSp is the number of siblings or spouses aborad.
Parch is the number of parents and chrildren aboard.

# Building & Training Model

```python
from sklearn.ensemble import RandomForestClassifier
 

X = train[["Pclass", "Sex", "Age", "SibSp", "Parch", "Embarked"]]  # Features
Y = train["Survived"]  # Lables

clf = RandomForestClassifier(n_estimators=99, criterion='gini', max_features='auto',
                             min_impurity_decrease=0.02, bootstrap=True, oob_score=True,
                             n_jobs=2, verbose=1, warm_start=False,
                             ccp_alpha=0.0, max_samples=None)
clf = clf.fit(X, Y)
```
